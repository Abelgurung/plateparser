```python
from functools import reduce
from pathlib import Path

import pandas as pd
```

```python
byonoy_df = pd.read_excel('examples/tecan_infinite_example.xlsx')
# byonoy_df = pd.read_csv('examples/byonoy_absolute_example.csv')
byonoy_df
```

```python
df = byonoy_df.dropna()
```

```python
df
```

```python
def find_start_idx(df):
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            element = df.iloc[i, j]
            if element == "A":
                return [i, j]

    print("Couldn't find 'A' well row")
    return None

start_idx = find_start_idx(df)
```

```python
start_idx
```

```python
standardPlates = [(48, 72), (32, 48), (16, 24), (8, 12), (6, 8), (4, 6), (3, 4), (2, 3)]
```

```python
for i in standardPlates:
    try:
        idx = i[0] - start_idx[0] + 1
        if df.iloc[idx, start_idx[1]] == chr(64 + idx):
            plateSize = i
        break
    except IndexError:
        continue
```

```python
indices_list = []
#Sometimes the well header is not recognized so we need to account for two cases
if start_idx == [0,0]:
    for row in range(plateSize[0]):
        for col in range(start_idx[1]+1, plateSize[1]+1):
            indices_list.append((row, col))
else:
    for i in range(plateSize[0]):
        for j in range((plateSize[1])):
            row = start_idx[0] + i
            col = start_idx[0]+ j
            indices_list.append((row, col))
```

```python
len(indices_list)
```

```python
row = plateSize[0]
col = plateSize[1]
```

```python
row,col
```

```python
wells = []
for i in range(1, row + 1):
    for j in range(1, col + 1):
        row_letter = chr(64 + i)
        wells.append(f"{row_letter}{j}")

rows = list(range(1, row + 1)) * col
cols = list(range(1, col + 1)) * row
cols.sort()

values = [float(df.iloc[idx]) for idx in indices_list]

df_wells = pd.DataFrame({
    'row': rows,
    'column': cols,
    'well': wells,
    'values': values
})
```
